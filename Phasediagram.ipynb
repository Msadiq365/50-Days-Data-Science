```python
import matplotlib.pyplot as plt
import numpy as np

# Define the XRD pattern data
angles = np.array([34.72, 38.533, 40.25, 44.798, 65.126, 69.51])
intensities = np.array([376, 2687, 300, 2289, 841, 276])

# Generate a smooth curve by interpolating the data points
x = np.linspace(30, 75, 1000)
y = np.zeros_like(x)
for angle, intensity in zip(angles, intensities):
    y += intensity * np.exp(-((x - angle) ** 2) / 0.1)

# Create the XRD plot
plt.figure(figsize=(12, 6))
plt.plot(x, y, color='red', linewidth=2)
plt.fill_between(x, y, color='red', alpha=0.3)

# Annotate the peaks with 2 values and IMCs
imc_annotations = {
    34.72: 'Fe3O4',
    38.533: 'Cr2O3',
    40.25: 'Na2CrO4',
    44.798: 'CaCl',
    65.126: 'CaCl',
    69.51: 'MgCl2'
}

for angle, intensity in zip(angles, intensities):
    plt.text(angle, intensity + 50, f'{angle:.2f}\n{imc_annotations[angle]}', ha='center', va='bottom')

# Set labels and title
plt.xlabel('2, degrees')
plt.ylabel('Intensity, cps')
plt.title('XRD Analysis')

# Show the XRD plot
plt.grid(axis='y')
plt.show()

# Define the phases and their normalized intensities
phases = ['Fe3O4', 'Cr2O3', 'Na2CrO4', 'CaCl', 'CaCl', 'MgCl2']
normalized_intensities = [14.61, 100.00, 12.50, 77.05, 26.18, 6.50]

# Create a pie chart for phase distribution
plt.figure(figsize=(8, 8))
plt.pie(normalized_intensities, labels=phases, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title('Phase Distribution in XRD Analysis')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define the phases and their normalized intensities for each category

# SS304 base metal
phases_ss304 = ['Fe3O4', 'Cr2O3', 'Na2CrO4', 'CaCl']
normalized_intensities_ss304 = [34.67, 12.45, 2.5, 0.89]

# Zinc/SS304 joint
phases_joint = ['CaCl', 'MgCl2', 'ZnO', 'Fe2O3']
normalized_intensities_joint = [14.61, 100.00, 12.50, 77.05]

# Zinc base metal
phases_zinc = ['ZnO', 'Zn(OH)2', 'MgCl']
normalized_intensities_zinc = [26.18, 6.50, 20.0]  # Adjusted values

# Function to create a pie chart
def create_pie_chart(phases, intensities, title):
    plt.figure(figsize=(8, 8))
    plt.pie(intensities, labels=phases, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
    plt.title(title)
    plt.legend(phases, title="IMCs", loc="upper right")
    plt.show()

# Create pie charts
create_pie_chart(phases_ss304, normalized_intensities_ss304, 'Phase Distribution in SS304 Base Metal')
create_pie_chart(phases_joint, normalized_intensities_joint, 'Phase Distribution in Zinc/SS304 Joint')
create_pie_chart(phases_zinc, normalized_intensities_zinc, 'Phase Distribution in Zinc Base Metal')
```

```python
import matplotlib.pyplot as plt

# Define the phases and their normalized intensities for each category

# SS304 base metal
phases_ss304 = ['Fe3O4', 'Cr2O3', 'Na2CrO4', 'CaCl']
normalized_intensities_ss304 = [34.67, 12.45, 2.5, 0.89]

# Zinc/SS304 joint
phases_joint = ['CaCl', 'MgCl2', 'ZnO', 'Fe2O3']
normalized_intensities_joint = [14.61, 100.00, 12.50, 77.05]

# Zinc base metal
phases_zinc = ['ZnO', 'Zn(OH)2', 'MgCl']
normalized_intensities_zinc = [26.18, 6.50, 20.0]  # Adjusted values

# Function to create a pie chart
def create_pie_chart(phases, intensities, title):
    plt.figure(figsize=(8, 8))
    wedges, _ = plt.pie(intensities, startangle=140, colors=plt.cm.tab20.colors, wedgeprops=dict(width=0.3, edgecolor='w'))

    # Create legend
    plt.legend(wedges, phases, title="IMCs", loc="upper right", bbox_to_anchor=(1.3, 1))
    plt.title(title)
    plt.show()

# Create pie charts
create_pie_chart(phases_ss304, normalized_intensities_ss304, 'Phase Distribution in SS304 Base Metal')
create_pie_chart(phases_joint, normalized_intensities_joint, 'Phase Distribution in Zinc/SS304 Joint')
create_pie_chart(phases_zinc, normalized_intensities_zinc, 'Phase Distribution in Zinc Base Metal')
```
